```python
from sklearn.impute import SimpleImputer
import pandas as pd

# Conversion numérique
df = df.apply(pd.to_numeric, errors="coerce")

# Colonnes numériques
num_cols = df.select_dtypes(include=["float64", "int64"]).columns

# Supprimer colonnes totalement vides
df = df.dropna(axis=1, how="all")

# Mettre à jour num_cols
num_cols = df.select_dtypes(include=["float64", "int64"]).columns

# Imputation
if len(num_cols) > 0:
    imputer_num = SimpleImputer(strategy="median")

    df[num_cols] = pd.DataFrame(
        imputer_num.fit_transform(df[num_cols]),
        columns=num_cols,
        index=df.index
    )

print("Imputation terminée ✅")
```

```python
print(df.shape)
print(num_cols)
print(df[num_cols].isna().sum())
```

```python
from sklearn.impute import SimpleImputer

# Sélection des colonnes catégorielles
cat_cols = df.select_dtypes(include=["object"]).columns

if len(cat_cols) > 0:
    imputer_cat = SimpleImputer(strategy="most_frequent")
    df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])
else:
    print("Aucune colonne catégorielle à imputer")
```

```python
import numpy as np
# on remplace les valeurs bizarre par les na

df = df.replace(["?", "\t?", "unknown"], np.nan)
```
